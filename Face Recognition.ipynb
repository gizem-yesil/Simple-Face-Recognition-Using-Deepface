Şimdi görüntüleri, tahminlerle beraber, bir fonksiyon ile yazdıralım.

```python
def visualize_attribute(img,result,attributes=["age","gender","race","emotion"]):

  image=cv2.imread(img)
  image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

  text = ""
  for attribute in attributes:
      if attribute in result:
          prediction = result[attribute]
          text += f"{attribute}: {prediction}\n"
      else:
          text += f"{attribute} not found in result\n"


  font= cv2.FONT_HERSHEY_SIMPLEX
  font_scale=1.5
  color=(255,255,255)
  thickness=2
  shadow_color=(0,0,0)


  plt.imshow(image_rgb)
  plt.axis('off')
  plt.title(text)
  plt.show()
```

"os" modülü ile dosyamız içerisindeki tüm "png, jpg, jpeg" dosyalarını alalım.

```python
image_paths = [os.path.join(images_path, f) for f in os.listdir(images_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
```

```python
for path in image_paths:
  results=DeepFace.analyze(path,actions=['age','gender','race','emotion'])

  for result in results:
    visualize_attribute(path, result)
```

Tahminleri yaparken action'ları yazdırdığımızda gördüğümüz üzere farklı sınıflara ait istatistikler gelyiyor. Onun yerine en yüksek olasılıklı tahminleri yazdırabiliriz.

Yukarıda her bir action'i incelediğimizde bunları temsil eden "dominant_gender,dominant_race" gibi parametrelerin var olduğunu görebiliriz. Bu parametreleri "get" metodu ile alabiliriz.

```python
for path in image_paths:
  results = DeepFace.analyze(img_path=path, actions=['age', 'gender', 'race', 'emotion'])

  for i, result in enumerate(results, start=1):
    attributes = {
        "age": result.get("age"),
        "gender": result.get("dominant_gender"),
        "race": result.get("dominant_race"),
        "emotion": result.get("dominant_emotion")
        }

    visualize_attribute(path, attributes, attributes=list(attributes.keys()))
```
